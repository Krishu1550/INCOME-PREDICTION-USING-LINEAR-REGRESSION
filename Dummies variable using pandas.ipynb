```python
import pandas as pd
```

```python
df=pd.read_csv("Car.csv")
df
```

```python
df=df.iloc[:,:4]
```

```python
df
```

```python
d_df=pd.get_dummies(df['Car Model'])
d_df
```

```python
final=pd.concat([df,d_df],axis='columns')
final
```

```python
final=final.drop(['Car Model','Audi A5'],axis='columns')
```

```python
x=final.drop(['Sell Price($)'],axis='columns')
x
```

```python
y=final['Sell Price($)']
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.predict([[3457,5,0,0]])
```

```python
p_dic={"Mileage":[3457,5634,6532],
      "Age":[2,4,2],
      "BMW":[0,1,1],
      "Mercedes":[0,0,1]}
p_dic
```

```python
p_df=pd.DataFrame(p_dic)
p_df
```

```python
for i in range(len(p_df)):
    temp= reg.predict([[p_df.Mileage[i],p_df.Age[i],p_df.BMW[i],p_df.Mercedes[i]]])
    p_df["Predicted_Price"]=temp[0]
    
```

```python
p_df.drop('Price',axis=1,inplace=True)
```

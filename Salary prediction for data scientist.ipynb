```python
def revenue(obj):
    try:
        obj=obj.replace('$','')  
        rev1= obj.split('to')
        if('million' in rev1[0].lower()):
            prev1=rev1[0].split('m')
            prev1=int(prev1[0])
            prev1=prev1*1000000
        elif ('billion'in rev1[0].lower()):
            prev1=rev1[0].split('b')
            prev1=int(prev1[0])
            prev1=prev1*1000000000
        elif('million' in rev1[1].lower()):
            prev1=rev1[0]
            prev1=int(prev1)
            prev1=prev1*1000000
        elif('billion'in rev1[1].lower()):
            prev1=rev1[0]
            prev1=int(prev1)
            prev1=prev1*1000000000
        if('million' in rev1[1].lower()):
            prev2=rev1[1].split('m')
            prev2=int(prev2[0])
            prev2=prev2*1000000
        elif('billion'in rev1[1].lower()):
            prev2=rev1[1].split('b')
            prev2=int(prev2[0])
            prev2=prev2*1000000000
        return (prev1+prev2)/2
    except:
        obj=obj.replace('$','')
        
        if('+' in obj):
       
            revenue=obj.split('+')
            revenue=int(revenue[0])*1000000
            return(revenue)
        elif('Less than' in obj) :
            obj=obj.replace('Less than','')
            revenue=obj.split('m')
            revenue= int(revenue[0])*1000000
            return revenue
        else:
            return None
```

```python
df['Revenue (M)']=df['Revenue'].apply(revenue)
```

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'others'
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split (X,y,test_size=0.2)
```
